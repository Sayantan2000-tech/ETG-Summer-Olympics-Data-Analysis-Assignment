**Summer Olympics Data Analysis Assignment**

```python
#Importing data from the github link to the python notebook

import pandas as pd
import requests
url = 'https://raw.githubusercontent.com/AshishJangra27/Skill-India-AI-ML-Scholarship/main/Summer%20Olympics/summer.csv'
res = requests.get(url, allow_redirects=True)
with open('games.csv','wb') as file:
    file.write(res.content)
games = pd.read_csv('games.csv')
```

```python
games.head()
```

**1. In how many cities Summer Olympics is held so far?**

```python
a=games['City'].unique().size
print("Total number of cities Summer Olympics is held so far is",a)
```

**2. Which sport is having most number of Gold Medals so far? (Top 5)**

```python
sd=games.loc[:,['Sport','Medal']]
sd.drop(index=sd[sd['Medal']!='Gold'].index , inplace=True)
print(sd.value_counts().head())

data =[]
for brand in sd['Sport'].unique():
    data.append([brand , len(sd[sd['Sport']  == brand])])

pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).head().plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (5,5))
```

**3. Which sport is having most number of medals so far? (Top 5)**

```python
df=games.loc[:,['Sport','Medal']]
data = []

for brand in df['Sport'].unique():
    data.append([brand , len(df[df['Sport']  == brand])])
print(pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).head())
pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).head().plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (5,5))
```

**4. Which player has won most number of medals? (Top 5)**

```python
df=games.loc[:,['Athlete','Medal']]
data = []

for b in df['Athlete'].unique():
    data.append([b , len(df[df['Athlete']  == b])])
print(pd.DataFrame(data,columns = ['Athlete','Medal']).sort_values(by='Medal', ascending=False).head())
pd.DataFrame(data,columns = ['Athlete','Medal']).sort_values(by='Medal', ascending=False).head().plot(x = 'Athlete', y = 'Medal', kind = 'bar', figsize = (5,5))
```
